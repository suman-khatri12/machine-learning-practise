### Recursive feature elimiantion with random forest

```python
from sklearn.feature_selection import RFE
rfe = RFE(estimator= rand, n_features_to_select= 5, step = 1).fit(X_train, y_train)
```

### Recursive feature elimination with cross validation and random forest

```python
from sklearn.feature_selection import RFECV
rfecv = RFECV(estimator=rand, cv = 5, scoring = 'accuracy').fit(X_train, y_train)
```

```python
print("optimal number of features:", rfecv.n_features_)
print("Best features:",X_train.columns[rfecv.support_])
```

```python
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score of number of selected features")
plt.plot(range(1, len(rfecv.grid_scores_)+ 1), rfecv.grid_scores_)
plt.show()
```

### Tree based feature selection and random forest classificaion

```python
rand.fit(X_train, y_train)
importances = rand.feature_importances_
std = np.std([tree.feature_importances_ for tree in rand.estimators_], axis = 0)
indices = np.argsort(importances)[::-1]
```

```python
print("feature ranking")
for f in range(X_train.shape[1]):
    print("%d feature %d (%f)" %(f + 1, indices[f], importances[indices[f]]))

plt.figure(1, figsize=(14, 13))
plt.title("Feature Impotances")
plt.bar(range(X_train.shape[1]), importances[indices], color = 'g', yerr= std[indices], align = 'center')
plt.xticks(range(x_train.shape[1]), x_train.columns[indices],rotation = 90)
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

```python
X_train_n = pd.DataFrame(X_train_normalized, columns=columns)
X_test_n = pd.DataFrame(X_test_normalized, columns = columns)
from sklearn.decomposition import PCA
pca = PCA().fit(X_train_n)
```

```python
plt.figure(1, figsize=(14, 13))
plt.clf()
plt.axes([.2, .2, .7, .7])
plt.plot(pca.explained_variance_ratio_, linewidth = 2)
plt.axis('tight')
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_')
```
