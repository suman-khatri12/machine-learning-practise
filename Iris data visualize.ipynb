```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
data = pd.read_csv("E://Iris.csv")
data1 = data.copy()
```

```python
data.species
```

```python
data['species'] = data['species'].map({"setosa":0, "virginica":1, "versicolor":2})
```

```python
data
```

```python
sns.lineplot(x = data.sepal_length, y = data.sepal_width, hue=data1.species)
```

```python
sns.histplot(x = data.sepal_length, y = data.sepal_width, hue = data.species, kde = True)
```

## Scatterplot


```python
plt.scatter(x = data.sepal_length, y=data.sepal_width)
plt.xlabel = "Sepal Length"
plt.ylabel = "Sepal Width"
plt.show()
```

```python
sns.scatterplot(x = data.sepal_length, y = data.sepal_width, hue = data.species)
```

```python
sns.jointplot(x = data.sepal_length, y = data.sepal_width)
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = data.sepal_length, y = data.sepal_width, hue= data.species, palette = "Set1")
```

```python
sns.boxplot(x = data.species, y = data.petal_length)
```

```python
colors = ('red', 'green', 'black')
```

```python
def size(m, n):
    plt.figure(figsize = (m , n))
```

```python
size(10, 10)
ax = sns.boxplot(x = data.species , y = data.petal_length)
ax = sns.stripplot(x = data.species, y = data.petal_length, palette= colors)
```

```python
size(10, 10)
sns.violinplot(x = data.species, y = data.petal_length)
```

### using kdeplot for univariate relationship

```python
size(10, 10)
ax = sns.kdeplot(x = data['petal_length'] , hue = data.species, palette= colors)
ax = sns.kdeplot(x = data.petal_width , hue= data.species, palette= 'Set3')
```

```python
sns.pairplot(data[['petal_length']], hue =data.species)
```

```python
data.boxplot(by='species', figsize = (12, 10))
plt.show()
```
